```python
import json
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load data
with open('dataset/exam_data1.json', 'r') as f:
    exam_data = json.load(f)
exam_data[0]
```

```python
# 訓練データの読み込み
with open('dataset/train_q.json', 'r') as train_file:
    train_data = json.load(train_file)
    
# 辞書形式に変換
exam_data_dict = {item[0]: item[1] for item in exam_data}
```

```python
# 学習データの作成
titles = []
correct_ids = []
candidates = []
texts = []

for item in train_data:
    titles.append(item['title_id'])
    correct_ids.append(item['ans_id'])
    candidates.append(item['candidates'])
    
    candidates_text = []
    for candidate in item['candidates']:
        candidates_text.append(exam_data_dict[candidate])
    texts.append(' '.join(candidates_text))

# candidatesを文字列のリストに変換
texts[0]
```

```python
# 特徴量の抽出
vectorizer = TfidfVectorizer(ngram_range=(1, 2))
X_train = vectorizer.fit_transform(texts)
```

```python
# 正解ラベルの作成
y_train = []
for i in range(len(correct_ids)):
    y_train.append(correct_ids[i] in candidates[i])
```

```python
# モデルのトレーニング
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
with open('dataset/test_q.json', 'r') as test_file:
    test_data = json.load(test_file)
```

```python
from tqdm import tqdm

suggestion = []

for item in tqdm(test_data):
    title_id = item['title_id']
    test_candidates = item['candidates']

    test_texts = []
    for candidate in test_candidates:
        if candidate in exam_data_dict:
            test_texts.append(exam_data_dict[candidate])
    
    X_test = vectorizer.transform(test_texts)
    predictions = model.predict_proba(X_test) # 予測確率の取得

    # 予測確率の高い順に本文候補を並べ替え
    sorted_candidates = [x.replace(" ", "") for _, x in sorted(zip(predictions, test_candidates), reverse=True)]

    # 結果の表示
    suggestion.append({'title_id': title_id, 'candidates': sorted_candidates})


with open('dataset/suggestion.json', 'w') as suggestion_file:
    json.dump(suggestion, suggestion_file, indent=4, ensure_ascii=False)
```
